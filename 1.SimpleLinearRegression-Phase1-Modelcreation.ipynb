```python
# pandas is a built-in library function (moreover like a class) used to read file, tables & commonly used for data analysis
# And now after importing pandas all the functions inside pandas get loaded in our new python page memory
# pandas are assigned/converted to a short form as pd (namechange)
import pandas as pd
```

```python
# Our input values(dataset) are assigned to the variable "dataset"
# Pandas(classname).read_csv(function name) 
# "read_csv" command is used to read the input .csv file which should be mentioned in the parameter in quotes i.e input filename
dataset=pd.read_csv("Salary_Data.csv")
```

```python
# just as printstatement we can print the values in variable "dataset"
dataset
```

```python
# YearsExperience - independent variable (x) & Salary - dependent variable (y) 
# The column "YearsExperience" alone inside "dataset" variable is assigned to another variable "independent"
# If we want to assign more than 1 column, then we can use comma , to mention the n number of column names
# The column name should be mentioned within double squared paranthesis and double quotes [[""]]
independent=dataset[["YearsExperience"]]
```

```python
#Just as a print statement, we use independent variable name alone to print the values stored inside the variable
independent
```

```python
# same as independent variable, we are assigning the values in "salary" column alone to "dependent" variable
dependent=dataset[["Salary"]]
```

```python
dependent
```

```python
# procedure line
# from filename import classname,we are importing the particular class"train_test_splits"(less memory usage as particular import)
# from Machine Language library"sklearn"(folder)inside "model_selection"(folder have to import the class "train-test_splits")
from sklearn.model_selection import train_test_splits
```

```python
# procedure line
# R.H.S = calling the class/function "train_test_split" - input parameters are included as independent, dependent
# test_size is almost always training set size will be 70% & test test size will be 30%
# random_state will be explained later by mam in another video
# L.H.S = This particular function is defined with 4 output values so we can use in another statement as input
# as per logic, independent variable (input) values are split into training set(70%)and testing set (30%) 
# so obviously dependent variable (output) also split into training set (70%) and testing set (30%)
# so we are using 4 variables to store the input and output values
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
# procedure lines #creation of model

# from "sklearn" library folder-inside folder "linear_model"  importing "LinearRegression" class
from sklearn.linear_model import LinearRegression
# model variable "regressor" is assigned with all functions in the class "LinearRegression" without parameter
regressor=LinearRegression()
# in regressor class, fit function is called with the input parameter(material) "X_train and Y_train"
# here weight(slope) and intercept(bias/initial value) will be calculated and substituted--> y=mx+c 
# and now its ready to calculate with input values
# model created 
regressor.fit(X_train,Y_train)
```

```python
# To check the value of weight/slope
# Co-efficient method is used to calculate the slope
# the value of slope is stored in the variable "weight"
weight=regressor.coef_
```

```python
# Like printout statement 
weight
```

```python
# To check the value of bias/origin/intercept
# Intercept method is used to calculate the intercept/original value
# The value is stored in variable "bias"
bias=regressor.intercept_
```

```python
# Like printout statement to display the value
bias
```

```python
# Evaluation (To cross check whether AI brain learned well)
# Now test set value get enters the AI brain and cross checking happens and brain predicts the output
# L.H.S = Y_pred is a variable created
# R.H.S = as we know regressor variable is assigned with all functions of linear regression;  
#predict is a function; Evaluation should be done with the test dataset X_test;
Y_pred=regressor.predict(X_test)
```

```python
# To find r^2 value
# sklearn is the library in which metrics function/class (folder) is called & importing r2_score function/class
from sklearn.metrics import r2_score
```

```python
#r_score is a variable to store output
# through r2_score function we are passing "Y_test" test dataset values and "Y_pred" values as parameters 
r_score=r2_score(Y_test,Y_pred)
```

```python
# To display the output value of r_score
# Since the output value is close to 1, our AI brain created good model 
# if it is close to 0, our AI brain created model is not a good one
r_score
```

```python
# To save a model
# importing pickle function

import pickle

# final filename can be stored in "filename" variable
# R.H.S = "finalized model (type of algorithm say linear)"
           #for pickle, the extension should be "".sav"
    
filename="finalized_model_linear.sav"

# filename alone created and model is not yet assigned
```

```python
# To assign values
#'wb' - Write binary; rb-read binary

pickle.dump(regressor,open(filename,'wb'))
```

```python
# To load/read the saved model
# Check in the jupyterwhether "finalized_model_linear.sav" is saved as well as in the folder
# that should be assigned to"loaded_model" & predict function is used to predict the final output and stored in result variable
# rb-read binary (to read only)
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[15]])
```
